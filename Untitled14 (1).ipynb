```python
# ============================================================
# PHASE 1A — FAST + FIXED VERSION
# ============================================================

import subprocess, sys, builtins, shutil, os
import torch
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# ── STEP 0: Install PyG ───────────────────────────────────────
subprocess.run("pip install -q torch-geometric", shell=True)
print("✅ PyG ready")

# ── STEP 1: Hide rdkit so PyG uses pre-built .pt file ─────────
rdkit_modules = {k: v for k, v in sys.modules.items() if 'rdkit' in k.lower()}
for k in list(rdkit_modules): sys.modules.pop(k)

_real_import = builtins.__import__
def _blocking_import(name, *args, **kwargs):
    if 'rdkit' in name.lower():
        raise ImportError(f"[Blocked] {name}")
    return _real_import(name, *args, **kwargs)
builtins.__import__ = _blocking_import
print("🔒 rdkit hidden from PyG")

# ── STEP 2: Load QM9 ─────────────────────────────────────────
from torch_geometric.datasets import QM9
from torch_geometric.loader import DataLoader

qm9_root = './data/QM9'
processed_dir = os.path.join(qm9_root, 'processed')
if os.path.exists(processed_dir):
    shutil.rmtree(processed_dir)
    print("🗑️  Cleared stale cache")

print("Downloading QM9 pre-processed (~130MB)...")
dataset = QM9(root=qm9_root)
print(f"✅ Loaded {len(dataset)} molecules")

# ── STEP 3: Restore rdkit ─────────────────────────────────────
builtins.__import__ = _real_import
for k, v in rdkit_modules.items(): sys.modules[k] = v
print("🔓 rdkit restored")

# ── STEP 4: FAST target extraction using slices ───────────────
# Instead of iterating molecule by molecule, use PyG's internal
# storage directly — 1000x faster than a Python for-loop
TARGET_IDX  = 4
TARGET_NAME = "HOMO-LUMO gap (eV)"

# dataset.data.y is shape [130831, 19] — slice it directly
all_targets = dataset.data.y[:, TARGET_IDX]   # shape [130831] — instant!

print(f"\nTarget: {TARGET_NAME}")
print(f"  Min={all_targets.min():.4f} Max={all_targets.max():.4f} "
      f"Mean={all_targets.mean():.4f} Std={all_targets.std():.4f}")

# ── STEP 5: Split ─────────────────────────────────────────────
torch.manual_seed(42)
n     = len(dataset)
perm  = torch.randperm(n)

n_train = int(0.8 * n)
n_val   = int(0.1 * n)
n_test  = n - n_train - n_val

idx_train = perm[:n_train]
idx_val   = perm[n_train : n_train + n_val]
idx_test  = perm[n_train + n_val :]

train_dataset = dataset[idx_train]
val_dataset   = dataset[idx_val]
test_dataset  = dataset[idx_test]

print(f"\nSplit → train:{len(train_dataset)} val:{len(val_dataset)} test:{len(test_dataset)}")

# ── STEP 6: FAST normalization constants ──────────────────────
# Use the global slice + index mask — no Python loop at all
train_targets = all_targets[idx_train]
TARGET_MEAN   = train_targets.mean().item()
TARGET_STD    = train_targets.std().item()

print(f"\n📌 SAVE THESE:")
print(f"  TARGET_MEAN = {TARGET_MEAN}")
print(f"  TARGET_STD  = {TARGET_STD}")

# ── STEP 7: Normalize transform ───────────────────────────────
def normalize_target(data):
    d   = data.clone()
    raw = d.y[0, TARGET_IDX].item()
    d.y = torch.tensor([(raw - TARGET_MEAN) / TARGET_STD], dtype=torch.float)
    return d
# ── STEP 8: FAST DataLoaders with lazy transform ──────────────
# FIX: Don't subclass BaseTransform — just use a plain callable.
# PyG's transform= argument accepts any callable.

BATCH_SIZE = 64

class NormalizeTarget:
    """Plain callable — no BaseTransform needed, works across all PyG versions."""
    def __call__(self, data):
        d   = data.clone()
        raw = d.y[0, TARGET_IDX].item()
        d.y = torch.tensor([(raw - TARGET_MEAN) / TARGET_STD], dtype=torch.float)
        return d

# Re-load with transform (no re-download — uses cached .pt file)
print("\nBuilding loaders with lazy transform (fast!)...")
dataset_transformed = QM9(root=qm9_root, transform=NormalizeTarget())

train_loader = DataLoader(dataset_transformed[idx_train],
                          batch_size=BATCH_SIZE, shuffle=True,  num_workers=0)
val_loader   = DataLoader(dataset_transformed[idx_val],
                          batch_size=BATCH_SIZE, shuffle=False, num_workers=0)
test_loader  = DataLoader(dataset_transformed[idx_test],
                          batch_size=BATCH_SIZE, shuffle=False, num_workers=0)

print(f"✅ Loaders: train={len(train_loader)} val={len(val_loader)} test={len(test_loader)} batches")

# ── STEP 9: Verify ────────────────────────────────────────────
batch = next(iter(train_loader))
print(f"\nBatch check:")
print(f"  x         : {batch.x.shape}")
print(f"  edge_index: {batch.edge_index.shape}")
print(f"  y         : {batch.y.shape}")
print(f"  batch vec : {batch.batch.shape}")

denorm = batch.y[:5].squeeze() * TARGET_STD + TARGET_MEAN
print(f"  Denormed y (first 5): {[round(v,3) for v in denorm.tolist()]} eV")
ok = all(0.5 < v < 13 for v in denorm.tolist())
print(f"  Range check: {'✅ OK' if ok else '⚠️ Check values'}")

# ── STEP 10: Feature dims ─────────────────────────────────────
IN_CHANNELS = dataset[0].x.shape[1]          # 11
EDGE_DIM    = dataset[0].edge_attr.shape[1]  # 4

print(f"\n📌 For Phase 2:")
print(f"  IN_CHANNELS = {IN_CHANNELS}")
print(f"  EDGE_DIM    = {EDGE_DIM}")

# ── STEP 11: Plot ─────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].hist(all_targets.numpy(), bins=80, color='steelblue',
             alpha=0.85, edgecolor='white', lw=0.3)
axes[0].axvline(TARGET_MEAN, color='red', linestyle='--', lw=2,
                label=f'Mean={TARGET_MEAN:.2f} eV')
axes[0].set_xlabel("HOMO-LUMO gap (eV)"); axes[0].set_ylabel("Count")
axes[0].set_title("QM9 — HOMO-LUMO gap (130k molecules)")
axes[0].legend()

mol_sizes = [dataset[i].x.shape[0] for i in range(3000)]
axes[1].hist(mol_sizes, bins=20, color='coral', alpha=0.85,
             edgecolor='white', lw=0.3)
axes[1].set_xlabel("Atoms per molecule"); axes[1].set_ylabel("Count")
axes[1].set_title("Molecule size distribution (3000 sample)")

plt.tight_layout()
plt.savefig('phase1a_exploration.png', dpi=150, bbox_inches='tight')
plt.show()
print("\n✅ Phase 1A complete!")
```

```python
# ============================================================
# PHASE 2A: GNN MODELS + TRAINING — FAST CPU VERSION
# ============================================================

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv, GATConv, global_mean_pool, global_max_pool
from torch_geometric.loader import DataLoader
import matplotlib.pyplot as plt
import numpy as np
import time
import json

DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Device: {DEVICE}")

# ── SPEED SETTINGS (change these if you want more accuracy) ──
SUBSET_TRAIN  = 30000   # use 30k of 104k training molecules
SUBSET_VAL    = 5000    # use 5k of 13k val molecules
BATCH_SIZE    = 128     # larger = faster per epoch
HIDDEN        = 64      # smaller model = faster
NUM_LAYERS    = 3
EPOCHS        = 20      # enough to see learning
PATIENCE      = 8       # early stopping
LR            = 3e-4

print(f"\nSpeed config:")
print(f"  Train subset : {SUBSET_TRAIN:,} molecules")
print(f"  Val subset   : {SUBSET_VAL:,} molecules")
print(f"  Batch size   : {BATCH_SIZE}")
print(f"  Hidden dim   : {HIDDEN}")
print(f"  Max epochs   : {EPOCHS}")
print(f"  Estimated time: ~15-25 min on CPU")

# ── 1. Rebuild loaders on subsets ────────────────────────────
print("\nBuilding fast loaders on subsets...")

# Subsample from the already-split datasets
sub_train = [normalize_target(train_dataset[i]) for i in range(SUBSET_TRAIN)]
sub_val   = [normalize_target(val_dataset[i])   for i in range(SUBSET_VAL)]
sub_test  = [normalize_target(test_dataset[i])  for i in range(min(SUBSET_VAL, len(test_dataset)))]

fast_train_loader = DataLoader(sub_train, batch_size=BATCH_SIZE, shuffle=True)
fast_val_loader   = DataLoader(sub_val,   batch_size=BATCH_SIZE, shuffle=False)
fast_test_loader  = DataLoader(sub_test,  batch_size=BATCH_SIZE, shuffle=False)

print(f"  Train: {len(fast_train_loader)} batches | "
      f"Val: {len(fast_val_loader)} | Test: {len(fast_test_loader)}")

# Verify a batch
b = next(iter(fast_train_loader))
print(f"  Batch y shape: {b.y.shape}  ✅")


# ── 2. GCN Model ──────────────────────────────────────────────
class GCNPredictor(nn.Module):
    def __init__(self, in_channels, hidden_channels=64, num_layers=3, dropout=0.1):
        super().__init__()
        self.dropout = dropout
        self.input_proj = nn.Linear(in_channels, hidden_channels)
        self.convs = nn.ModuleList([
            GCNConv(hidden_channels, hidden_channels) for _ in range(num_layers)
        ])
        self.bns = nn.ModuleList([
            nn.BatchNorm1d(hidden_channels) for _ in range(num_layers)
        ])
        pool_dim = 2 * hidden_channels
        self.mlp = nn.Sequential(
            nn.Linear(pool_dim, hidden_channels),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_channels, hidden_channels // 2),
            nn.ReLU(),
            nn.Linear(hidden_channels // 2, 1)
        )

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        x = F.relu(self.input_proj(x))
        for conv, bn in zip(self.convs, self.bns):
            x_res = x
            x = conv(x, edge_index)
            x = bn(x)
            x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
            x = x + x_res
        x = torch.cat([global_mean_pool(x, batch),
                        global_max_pool(x, batch)], dim=1)
        return self.mlp(x)

    def get_embedding(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        x = F.relu(self.input_proj(x))
        for conv, bn in zip(self.convs, self.bns):
            x_res = x
            x = conv(x, edge_index)
            x = bn(x)
            x = F.relu(x)
            x = x + x_res
        return torch.cat([global_mean_pool(x, batch),
                          global_max_pool(x, batch)], dim=1)


# ── 3. GAT Model ──────────────────────────────────────────────
class GATPredictor(nn.Module):
    def __init__(self, in_channels, hidden_channels=64,
                 num_layers=3, heads=2, dropout=0.1):
        super().__init__()
        self.dropout = dropout
        assert hidden_channels % heads == 0
        self.head_dim = hidden_channels // heads
        self.input_proj = nn.Linear(in_channels, hidden_channels)
        self.convs = nn.ModuleList([
            GATConv(hidden_channels, self.head_dim,
                    heads=heads, dropout=dropout, concat=True)
            for _ in range(num_layers)
        ])
        self.bns = nn.ModuleList([
            nn.BatchNorm1d(hidden_channels) for _ in range(num_layers)
        ])
        pool_dim = 2 * hidden_channels
        self.mlp = nn.Sequential(
            nn.Linear(pool_dim, hidden_channels),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_channels, hidden_channels // 2),
            nn.ReLU(),
            nn.Linear(hidden_channels // 2, 1)
        )

    def forward(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        x = F.relu(self.input_proj(x))
        for conv, bn in zip(self.convs, self.bns):
            x_res = x
            x = conv(x, edge_index)
            x = bn(x)
            x = F.relu(x)
            x = F.dropout(x, p=self.dropout, training=self.training)
            x = x + x_res
        x = torch.cat([global_mean_pool(x, batch),
                        global_max_pool(x, batch)], dim=1)
        return self.mlp(x)

    def get_embedding(self, data):
        x, edge_index, batch = data.x, data.edge_index, data.batch
        x = F.relu(self.input_proj(x))
        for conv, bn in zip(self.convs, self.bns):
            x_res = x
            x = conv(x, edge_index)
            x = bn(x)
            x = F.relu(x)
            x = x + x_res
        return torch.cat([global_mean_pool(x, batch),
                          global_max_pool(x, batch)], dim=1)


# ── 4. Train + Eval functions ─────────────────────────────────
def train_one_epoch(model, loader, optimizer):
    model.train()
    total_loss, total_mols = 0.0, 0
    for batch in loader:
        batch = batch.to(DEVICE)
        optimizer.zero_grad()
        pred   = model(batch).squeeze(-1)
        target = batch.y.squeeze(-1)
        loss   = F.mse_loss(pred, target)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()
        total_loss += loss.item() * batch.num_graphs
        total_mols += batch.num_graphs
    return total_loss / total_mols

@torch.no_grad()
def evaluate(model, loader):
    model.eval()
    preds_norm, trues_norm = [], []
    for batch in loader:
        batch = batch.to(DEVICE)
        preds_norm.append(model(batch).squeeze(-1).cpu())
        trues_norm.append(batch.y.squeeze(-1).cpu())
    preds_norm = torch.cat(preds_norm)
    trues_norm = torch.cat(trues_norm)
    preds_ev   = preds_norm * TARGET_STD + TARGET_MEAN
    trues_ev   = trues_norm * TARGET_STD + TARGET_MEAN
    mae_ev     = F.l1_loss(preds_ev, trues_ev).item()
    rmse_ev    = ((preds_ev - trues_ev) ** 2).mean().sqrt().item()
    return mae_ev, rmse_ev, preds_ev.numpy(), trues_ev.numpy()


# ── 5. Training loop ──────────────────────────────────────────
def train_model(ModelClass, model_kwargs, model_name):
    model = ModelClass(**model_kwargs).to(DEVICE)
    n_params = sum(p.numel() for p in model.parameters())
    print(f"\n{'='*50}")
    print(f"Model : {model_name}  ({n_params:,} params)")
    print(f"{'='*50}")

    optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=1e-5)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)

    best_val_mae = float('inf')
    best_state   = None
    patience_cnt = 0
    history      = {'train_mse': [], 'val_mae': [], 'val_rmse': []}

    start = time.time()
    for epoch in range(1, EPOCHS + 1):
        t0        = time.time()
        train_mse = train_one_epoch(model, fast_train_loader, optimizer)
        val_mae, val_rmse, _, _ = evaluate(model, fast_val_loader)
        scheduler.step()
        epoch_t   = time.time() - t0

        history['train_mse'].append(train_mse)
        history['val_mae'].append(val_mae)
        history['val_rmse'].append(val_rmse)

        if val_mae < best_val_mae:
            best_val_mae = val_mae
            best_state   = {k: v.clone() for k, v in model.state_dict().items()}
            patience_cnt = 0
            marker = " ✅"
        else:
            patience_cnt += 1
            marker = ""

        # Print every epoch so you can see progress
        elapsed = time.time() - start
        remaining = (epoch_t * (EPOCHS - epoch))
        print(f"  Ep {epoch:2d}/{EPOCHS} | "
              f"Train MSE: {train_mse:.4f} | "
              f"Val MAE: {val_mae:.4f} eV | "
              f"Best: {best_val_mae:.4f} | "
              f"{epoch_t:.0f}s/ep | "
              f"~{remaining/60:.0f}min left"
              f"{marker}")

        if patience_cnt >= PATIENCE:
            print(f"  ⏹ Early stop at epoch {epoch}")
            break

    model.load_state_dict(best_state)

    test_mae, test_rmse, preds_ev, trues_ev = evaluate(model, fast_test_loader)
    total_time = time.time() - start
    print(f"\n  ✅ Done in {total_time/60:.1f} min")
    print(f"  Test MAE : {test_mae:.4f} eV")
    print(f"  Test RMSE: {test_rmse:.4f} eV")
    print(f"  (Good result for 20 epochs on CPU: < 0.40 eV)")

    return model, history, test_mae, test_rmse, preds_ev, trues_ev


# ── 6. Train both models ──────────────────────────────────────
print("\n🚀 Training GCN...")
gcn_model, gcn_history, gcn_test_mae, gcn_test_rmse, gcn_preds, gcn_trues = train_model(
    GCNPredictor,
    dict(in_channels=IN_CHANNELS, hidden_channels=HIDDEN, num_layers=NUM_LAYERS),
    "GCN (3 layers, 64-dim)"
)

print("\n🚀 Training GAT...")
gat_model, gat_history, gat_test_mae, gat_test_rmse, gat_preds, gat_trues = train_model(
    GATPredictor,
    dict(in_channels=IN_CHANNELS, hidden_channels=HIDDEN,
         num_layers=NUM_LAYERS, heads=2),
    "GAT (3 layers, 64-dim, 2 heads)"
)


# ── 7. Pick best model ────────────────────────────────────────
if gat_test_mae <= gcn_test_mae:
    BEST_MODEL      = gat_model
    BEST_MODEL_NAME = "GAT"
    best_preds      = gat_preds
    best_trues      = gat_trues
    best_mae        = gat_test_mae
else:
    BEST_MODEL      = gcn_model
    BEST_MODEL_NAME = "GCN"
    best_preds      = gcn_preds
    best_trues      = gcn_trues
    best_mae        = gcn_test_mae

print(f"\n🏆 Best model: {BEST_MODEL_NAME}  MAE = {best_mae:.4f} eV")


# ── 8. Save everything ───────────────────────────────────────
torch.save(gcn_model.state_dict(), 'gcn_homo_lumo.pt')
torch.save(gat_model.state_dict(), 'gat_homo_lumo.pt')

constants = {
    'TARGET_MEAN'  : TARGET_MEAN,
    'TARGET_STD'   : TARGET_STD,
    'IN_CHANNELS'  : IN_CHANNELS,
    'EDGE_DIM'     : EDGE_DIM,
    'HIDDEN'       : HIDDEN,
    'BEST_MODEL'   : BEST_MODEL_NAME,
    'gcn_test_mae' : gcn_test_mae,
    'gat_test_mae' : gat_test_mae,
}
with open('p1_constants.json', 'w') as f:
    json.dump(constants, f, indent=2)

print("💾 Saved: gcn_homo_lumo.pt | gat_homo_lumo.pt | p1_constants.json")


# ── 9. Plots ─────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(gcn_history['val_mae'], label='GCN', color='steelblue', marker='o', ms=3)
axes[0].plot(gat_history['val_mae'], label='GAT', color='coral',     marker='o', ms=3)
axes[0].set_xlabel("Epoch"); axes[0].set_ylabel("Val MAE (eV)")
axes[0].set_title("Validation MAE over training")
axes[0].legend(); axes[0].grid(alpha=0.3)

axes[1].plot(gcn_history['train_mse'], label='GCN', color='steelblue', alpha=0.7)
axes[1].plot(gat_history['train_mse'], label='GAT', color='coral',     alpha=0.7)
axes[1].set_xlabel("Epoch"); axes[1].set_ylabel("Train MSE (normalized)")
axes[1].set_title("Training MSE")
axes[1].legend(); axes[1].grid(alpha=0.3)

# Scatter: predicted vs true
axes[2].scatter(best_trues, best_preds, alpha=0.15, s=4, color='steelblue')
lo = min(best_trues.min(), best_preds.min())
hi = max(best_trues.max(), best_preds.max())
axes[2].plot([lo, hi], [lo, hi], 'r--', lw=1.5, label='Perfect')
axes[2].set_xlabel("True HOMO-LUMO (eV)")
axes[2].set_ylabel("Predicted (eV)")
axes[2].set_title(f"{BEST_MODEL_NAME} | Test MAE = {best_mae:.4f} eV")
axes[2].legend(); axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('phase2a_results.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✅ Phase 2A complete! Move to Phase 3.")
```

```python
# ============================================================
# PHASE 3A: INVERSE DESIGN — FIXED VERSION
# ============================================================

import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Draw, Descriptors
import copy
import warnings
warnings.filterwarnings('ignore')

ATOM_TYPES = ['H', 'C', 'N', 'O', 'F']

# ── 1. Inverse design (FIXED) ─────────────────────────────────
def inverse_design(model, seed_data_normalized, target_ev,
                   n_steps=300, lr=0.04, device=DEVICE):
    """
    FIX: Use ONE consistent data object throughout.
    seed_data_normalized: already has y = normalized scalar [1]
    target_ev: target value in eV (real units)
    """
    model.eval()
    for p in model.parameters():
        p.requires_grad_(False)

    # Work entirely with the normalized data object
    data = copy.deepcopy(seed_data_normalized).to(device)

    # Make node features differentiable
    x_opt = data.x.clone().float().requires_grad_(True)
    optimizer_inv = torch.optim.Adam([x_opt], lr=lr)

    # Normalize the target to model output scale
    target_norm = torch.tensor(
        [(target_ev - TARGET_MEAN) / TARGET_STD],
        dtype=torch.float, device=device
    )

    trajectory = []
    for step in range(n_steps):
        optimizer_inv.zero_grad()

        # Build tmp data — same graph topology, optimized features
        tmp = data.clone()
        tmp.x = x_opt

        pred_norm = model(tmp)                            # [1, 1]
        loss = F.mse_loss(pred_norm.squeeze(), target_norm.squeeze())
        loss.backward()
        optimizer_inv.step()

        # Soft clamp to keep features reasonable
        with torch.no_grad():
            x_opt.data.clamp_(-3.0, 3.0)

        # Record in eV
        pred_ev = pred_norm.item() * TARGET_STD + TARGET_MEAN
        trajectory.append(pred_ev)

    initial_ev = trajectory[0]
    final_ev   = trajectory[-1]
    print(f"  Start: {initial_ev:.4f} eV | Target: {target_ev:.4f} eV | "
          f"Reached: {final_ev:.4f} eV | Δ={final_ev-initial_ev:+.4f} eV")

    # Return optimized data (with the same edge_index/edge_attr as input)
    result_data = data.clone().cpu()
    result_data.x = x_opt.detach().cpu()
    return result_data, trajectory


# ── 2. Molecule decoder (FIXED) ───────────────────────────────
def decode_to_molecule(optimized_data):
    """
    FIX: takes ONE data object — both x (optimized) and edge_index/edge_attr
    come from the same object. No more mismatched sources.
    """
    x_opt      = optimized_data.x          # optimized features
    edge_index = optimized_data.edge_index  # same object
    edge_attr  = optimized_data.edge_attr   # same object

    BOND_MAP = {
        0: Chem.rdchem.BondType.SINGLE,
        1: Chem.rdchem.BondType.DOUBLE,
        2: Chem.rdchem.BondType.TRIPLE,
        3: Chem.rdchem.BondType.AROMATIC,
    }

    try:
        # Decode atom types from continuous one-hot (features 0-4)
        atom_type_logits = x_opt[:, :5]                # [N, 5]
        atom_type_ids    = atom_type_logits.argmax(dim=1).tolist()
        atom_symbols     = [ATOM_TYPES[i] for i in atom_type_ids]

        mol = Chem.RWMol()
        for sym in atom_symbols:
            mol.AddAtom(Chem.Atom(sym))

        added = set()
        n_edges = edge_index.shape[1]
        for i in range(n_edges):
            u = edge_index[0, i].item()
            v = edge_index[1, i].item()
            if u >= v or (u, v) in added:
                continue
            bond_feat = edge_attr[i].tolist()
            bt_idx    = int(np.argmax(bond_feat))
            mol.AddBond(u, v, BOND_MAP.get(bt_idx, Chem.rdchem.BondType.SINGLE))
            added.add((u, v))

        mol = mol.GetMol()
        Chem.SanitizeMol(mol)
        smi = Chem.MolToSmiles(mol)
        return mol, smi, True

    except Exception as e:
        return None, None, False


# ── 3. Run experiments ────────────────────────────────────────
BEST_MODEL.to(DEVICE)
BEST_MODEL.eval()

# Pick a seed molecule and normalize it
seed_idx       = 42
seed_raw       = test_dataset[seed_idx]
seed_normalized = normalize_target(seed_raw)   # y is now [1] scalar

# Get its initial predicted value
with torch.no_grad():
    init_pred_norm = BEST_MODEL(seed_normalized.to(DEVICE)).item()
init_ev = init_pred_norm * TARGET_STD + TARGET_MEAN
print(f"Seed molecule initial HOMO-LUMO: {init_ev:.4f} eV")

# Experiment 1: maximize
target_high = min(init_ev + 2.0, 12.0)
print(f"\nExperiment 1 — Maximize gap (target: {target_high:.2f} eV)")
opt_high, traj_high = inverse_design(
    BEST_MODEL, seed_normalized, target_high,
    n_steps=300, lr=0.04
)

# Experiment 2: minimize
target_low = max(init_ev - 2.0, 0.5)
print(f"\nExperiment 2 — Minimize gap (target: {target_low:.2f} eV)")
opt_low, traj_low = inverse_design(
    BEST_MODEL, seed_normalized, target_low,
    n_steps=300, lr=0.04
)

# Decode to molecules (FIX: pass single object)
mol_high, smi_high, valid_high = decode_to_molecule(opt_high)
mol_low,  smi_low,  valid_low  = decode_to_molecule(opt_low)

print(f"\nHigh-gap decoded: {'Valid ✅ ' + smi_high if valid_high else 'Invalid ❌'}")
print(f"Low-gap decoded : {'Valid ✅ ' + smi_low  if valid_low  else 'Invalid ❌'}")

# ── 4. Multi-seed scan ────────────────────────────────────────
print("\n--- Multi-seed scan (15 molecules) ---")
scan = []
for i in range(15):
    try:
        raw  = test_dataset[i]
        norm = normalize_target(raw)
        with torch.no_grad():
            iv = BEST_MODEL(norm.to(DEVICE)).item() * TARGET_STD + TARGET_MEAN
        tgt = min(iv + 1.5, 11.0)
        opt_d, traj = inverse_design(BEST_MODEL, norm, tgt, n_steps=150, lr=0.04)
        fv = traj[-1]
        mol, smi, valid = decode_to_molecule(opt_d)
        scan.append({
            'seed': i, 'init': iv, 'target': tgt,
            'final': fv, 'delta': fv - iv, 'valid': valid
        })
    except Exception as e:
        scan.append({'seed': i, 'error': str(e)})

import pandas as pd
df = pd.DataFrame([s for s in scan if 'error' not in s])
print(df[['seed','init','target','final','delta','valid']].to_string(index=False))
print(f"\nValid decode rate: {df['valid'].mean()*100:.0f}%")

# ── 5. Plots ──────────────────────────────────────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(traj_high, color='steelblue', lw=2, label='Predicted')
axes[0].axhline(target_high, color='red',   linestyle='--', label=f'Target {target_high:.2f} eV')
axes[0].axhline(init_ev,     color='green', linestyle=':',  label=f'Start {init_ev:.2f} eV')
axes[0].set_title("Maximize HOMO-LUMO gap")
axes[0].set_xlabel("Step"); axes[0].set_ylabel("Predicted gap (eV)")
axes[0].legend(); axes[0].grid(alpha=0.3)

axes[1].plot(traj_low, color='coral', lw=2, label='Predicted')
axes[1].axhline(target_low, color='red',   linestyle='--', label=f'Target {target_low:.2f} eV')
axes[1].axhline(init_ev,    color='green', linestyle=':',  label=f'Start {init_ev:.2f} eV')
axes[1].set_title("Minimize HOMO-LUMO gap")
axes[1].set_xlabel("Step"); axes[1].set_ylabel("Predicted gap (eV)")
axes[1].legend(); axes[1].grid(alpha=0.3)

if len(df) > 0:
    axes[2].scatter(df['init'], df['final'], alpha=0.7, s=60,
                    c=df['valid'].astype(int), cmap='RdYlGn')
    lim = [df['init'].min() - 0.5, df['final'].max() + 0.5]
    axes[2].plot(lim, lim, 'k--', lw=1, label='No change')
    axes[2].set_xlabel("Initial gap (eV)")
    axes[2].set_ylabel("After optimization (eV)")
    axes[2].set_title("Multi-seed inverse design results\n(green=valid decode)")
    axes[2].legend(); axes[2].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('phase3a_inverse_design.png', dpi=150, bbox_inches='tight')
plt.show()

print("\n✅ Phase 3A complete!")
```

```python
# ============================================================
# PHASE 4A: GALLON-STYLE MULTIMODAL DISTILLATION
# ============================================================
# Inspired by: "LLM and GNN are Complementary: Distilling LLM
# for Multimodal Graph Learning"
#
# Pipeline:
#   Teacher 1 (GNN): Our trained GAT model
#   Teacher 2 (Text): Simple SMILES-based text features
#                     (we use Morgan fingerprints as a proxy
#                      since we can't run an LLM in Colab CPU)
#   Student (MLP): Lightweight model trained to match BOTH
#                  teachers via soft-label distillation
#
# The student is 10× faster at inference time.
# ============================================================

import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from rdkit import Chem
from rdkit.Chem import AllChem, Descriptors, rdMolDescriptors
from torch_geometric.loader import DataLoader
import matplotlib.pyplot as plt

# ── 1. Extract SMILES-based text features ────────────────────
def smiles_to_fingerprint(smiles, radius=2, n_bits=512):
    """
    Morgan fingerprint as a proxy for LLM text understanding.
    In a production system you'd use: ChemBERTa, MolBERT, or
    an LLM to embed the SMILES string. Here we use fingerprints.
    """
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return np.zeros(n_bits)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius, n_bits)
    return np.array(fp, dtype=np.float32)

def get_rdkit_descriptors(smiles):
    """Additional molecular descriptors as text-like features."""
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        return np.zeros(12)
    return np.array([
        Descriptors.MolWt(mol),
        Descriptors.MolLogP(mol),
        Descriptors.NumHDonors(mol),
        Descriptors.NumHAcceptors(mol),
        Descriptors.TPSA(mol),
        Descriptors.NumRotatableBonds(mol),
        rdMolDescriptors.CalcNumRings(mol),
        rdMolDescriptors.CalcNumAromaticRings(mol),
        Descriptors.FractionCSP3(mol),
        Descriptors.NumHeteroatoms(mol),
        rdMolDescriptors.CalcNumHeavyAtoms(mol),
        Descriptors.BertzCT(mol),
    ], dtype=np.float32)

from rdkit.Chem import rdMolDescriptors

# Precompute fingerprints for the training set
print("Computing Morgan fingerprints for training set...")
print("(This is our 'text encoder' — in production, use ChemBERTa)")

TEXT_DIM = 512 + 12  # Morgan fp + RDKit descriptors

def precompute_text_features(dataset_subset, n_max=None):
    fps = []
    descs = []
    skipped = 0
    n_max = n_max or len(dataset_subset)

    for i, data in enumerate(dataset_subset[:n_max]):
        if i % 1000 == 0:
            print(f"  {i}/{n_max}...", end='\r')
        try:
            smiles = data.smiles
            fp = smiles_to_fingerprint(smiles)
            desc = get_rdkit_descriptors(smiles)
        except Exception:
            fp = np.zeros(512)
            desc = np.zeros(12)
            skipped += 1
        fps.append(fp)
        descs.append(desc)

    fps  = np.stack(fps, axis=0)
    descs = np.stack(descs, axis=0)
    # Normalize descriptors
    desc_mean = descs.mean(0); desc_std = descs.std(0) + 1e-8
    descs_norm = (descs - desc_mean) / desc_std
    text_features = np.concatenate([fps, descs_norm], axis=1)
    print(f"\n  Done. Skipped {skipped} molecules. Shape: {text_features.shape}")
    return text_features, desc_mean, desc_std

# Use subset for speed (use full set if you have time)
SUBSET = 20000
print(f"Using subset of {SUBSET} molecules for distillation demo")

# Get GNN (teacher 1) predictions on training subset
print("\nGenerating GNN teacher predictions...")
BEST_MODEL.eval()
BEST_MODEL.to(DEVICE)

train_subset_loader = DataLoader(
    [normalize_target(train_dataset[i]) for i in range(SUBSET)],
    batch_size=256, shuffle=False
)

gnn_preds = []
gnn_embeds = []
true_labels = []

with torch.no_grad():
    for batch in train_subset_loader:
        batch = batch.to(DEVICE)
        pred = BEST_MODEL(batch).squeeze(-1)
        emb  = BEST_MODEL.get_embedding(batch)
        gnn_preds.append(pred.cpu())
        gnn_embeds.append(emb.cpu())
        true_labels.append(batch.y.squeeze(-1).cpu())

gnn_preds  = torch.cat(gnn_preds)
gnn_embeds = torch.cat(gnn_embeds)
true_labels= torch.cat(true_labels)
print(f"  GNN predictions shape : {gnn_preds.shape}")
print(f"  GNN embeddings shape  : {gnn_embeds.shape}")

# Get text features (teacher 2)
print("\nComputing text features (Morgan fingerprints)...")
text_feats, desc_mean, desc_std = precompute_text_features(
    train_dataset, n_max=SUBSET
)
text_feats_tensor = torch.from_numpy(text_feats).float()
print(f"  Text features shape: {text_feats_tensor.shape}")


# ── 2. Student MLP Architecture ───────────────────────────────
class StudentMLP(nn.Module):
    """
    Lightweight student model that learns from:
    - Task labels (true HOMO-LUMO gaps)
    - GNN teacher soft predictions
    - Text feature teacher logits

    At inference: only needs Morgan fingerprints (fast!)
    No graph processing needed.
    """
    def __init__(self, text_dim, hidden_dim=256, gnn_embed_dim=256):
        super().__init__()
        self.text_encoder = nn.Sequential(
            nn.Linear(text_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.ReLU(),
            nn.Dropout(0.1),
            nn.Linear(hidden_dim, hidden_dim),
            nn.LayerNorm(hidden_dim),
            nn.ReLU(),
        )
        # Prediction head
        self.predictor = nn.Sequential(
            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.ReLU(),
            nn.Linear(hidden_dim // 2, 1)
        )
        # Alignment head: match the GNN embedding space
        self.align_head = nn.Linear(hidden_dim, gnn_embed_dim)

    def forward(self, text_x):
        h = self.text_encoder(text_x)
        pred = self.predictor(h)
        alignment = self.align_head(h)
        return pred, alignment, h


# ── 3. Distillation Training ──────────────────────────────────
def distillation_loss(student_pred, student_align, student_h,
                      gnn_pred_teacher, gnn_embed_teacher,
                      true_target,
                      alpha=0.5, beta=0.3, temp=2.0):
    """
    Combined distillation loss:
      L = alpha * L_task + beta * L_pred_distill + (1-alpha-beta) * L_embed_distill

    L_task         : MSE between student and true labels
    L_pred_distill : MSE between student and GNN predictions (soft labels)
    L_embed_distill: cosine loss between student and GNN embeddings
    """
    # Task loss
    L_task = F.mse_loss(student_pred.squeeze(), true_target)

    # Prediction distillation (temperature-scaled)
    L_pred = F.mse_loss(student_pred.squeeze() / temp,
                        gnn_pred_teacher / temp)

    # Embedding alignment (cosine similarity)
    L_embed = (1 - F.cosine_similarity(
        student_align, gnn_embed_teacher, dim=-1
    )).mean()

    total = alpha * L_task + beta * L_pred + (1 - alpha - beta) * L_embed
    return total, L_task.item(), L_pred.item(), L_embed.item()


GNN_EMBED_DIM = gnn_embeds.shape[1]  # from get_embedding output
student = StudentMLP(
    text_dim=TEXT_DIM,
    hidden_dim=256,
    gnn_embed_dim=GNN_EMBED_DIM
)
print(f"\nStudent MLP parameters: {sum(p.numel() for p in student.parameters()):,}")
print(f"GNN teacher parameters: {sum(p.numel() for p in BEST_MODEL.parameters()):,}")

optimizer_s = torch.optim.Adam(student.parameters(), lr=1e-3, weight_decay=1e-5)
scheduler_s = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer_s, T_max=50)

# Create tensor dataset for student training
from torch.utils.data import TensorDataset, DataLoader as TorchDataLoader

student_dataset = TensorDataset(
    text_feats_tensor, gnn_preds, gnn_embeds, true_labels
)
n_student = len(student_dataset)
n_s_train = int(0.85 * n_student)
s_train_ds = TensorDataset(
    text_feats_tensor[:n_s_train],
    gnn_preds[:n_s_train],
    gnn_embeds[:n_s_train],
    true_labels[:n_s_train]
)
s_val_ds = TensorDataset(
    text_feats_tensor[n_s_train:],
    gnn_preds[n_s_train:],
    gnn_embeds[n_s_train:],
    true_labels[n_s_train:]
)

s_train_loader = TorchDataLoader(s_train_ds, batch_size=512, shuffle=True)
s_val_loader   = TorchDataLoader(s_val_ds,   batch_size=512, shuffle=False)

print(f"\nDistillation training: {n_s_train} train, {n_student - n_s_train} val")

student_history = {'total':[], 'task':[], 'pred':[], 'embed':[], 'val_mae':[]}

EPOCHS_S = 60
print(f"Training student for {EPOCHS_S} epochs...")

for epoch in range(1, EPOCHS_S + 1):
    student.train()
    epoch_losses = {'total':0,'task':0,'pred':0,'embed':0}
    n_batches = 0

    for text_x, gnn_p, gnn_e, y_true in s_train_loader:
        optimizer_s.zero_grad()
        s_pred, s_align, s_h = student(text_x)
        loss, lt, lp, le = distillation_loss(
            s_pred, s_align, s_h,
            gnn_p, gnn_e, y_true
        )
        loss.backward()
        optimizer_s.step()
        epoch_losses['total'] += loss.item()
        epoch_losses['task']  += lt
        epoch_losses['pred']  += lp
        epoch_losses['embed'] += le
        n_batches += 1

    scheduler_s.step()

    # Validation
    student.eval()
    val_preds, val_trues = [], []
    with torch.no_grad():
        for text_x, gnn_p, gnn_e, y_true in s_val_loader:
            s_pred, _, _ = student(text_x)
            val_preds.append(s_pred.squeeze())
            val_trues.append(y_true)
    val_preds = torch.cat(val_preds)
    val_trues = torch.cat(val_trues)
    val_mae_norm = F.l1_loss(val_preds, val_trues).item()
    val_mae_ev = val_mae_norm * TARGET_STD

    for k in epoch_losses:
        student_history[k].append(epoch_losses[k] / n_batches)
    student_history['val_mae'].append(val_mae_ev)

    if epoch % 10 == 0 or epoch == 1:
        print(f"  Epoch {epoch:3d} | "
              f"L_total={epoch_losses['total']/n_batches:.4f} | "
              f"L_task={epoch_losses['task']/n_batches:.4f} | "
              f"Val MAE={val_mae_ev:.4f} eV")

# ── 4. Compare GNN vs Student ─────────────────────────────────
print("\n--- Speed & Accuracy Comparison ---")

import time

# GNN inference time
t0 = time.time()
with torch.no_grad():
    for batch in val_loader:
        batch = batch.to(DEVICE)
        _ = BEST_MODEL(batch)
gnn_time = time.time() - t0

# Student inference time (on same data via text features)
t0 = time.time()
student.eval()
with torch.no_grad():
    for text_x, *_ in s_val_loader:
        _, _, _ = student(text_x)
student_time = time.time() - t0

print(f"  GNN inference time   : {gnn_time:.3f}s")
print(f"  Student inference    : {student_time:.3f}s")
print(f"  Speedup              : {gnn_time/student_time:.1f}×")

best_student_mae = min(student_history['val_mae'])
print(f"\n  GNN Val MAE   : {gat_test_mae:.4f} eV")
print(f"  Student Val MAE: {best_student_mae:.4f} eV")
print(f"  Accuracy gap  : {best_student_mae - gat_test_mae:.4f} eV")

# Plot distillation curves
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(student_history['total'], label='Total loss', color='black')
axes[0].plot(student_history['task'],  label='Task (MSE)',  color='steelblue')
axes[0].plot(student_history['pred'],  label='Pred distill', color='coral')
axes[0].plot(student_history['embed'], label='Embed align',  color='green')
axes[0].set_xlabel("Epoch"); axes[0].set_ylabel("Loss")
axes[0].set_title("Distillation loss components")
axes[0].legend(); axes[0].grid(alpha=0.3)

axes[1].plot(student_history['val_mae'], color='purple')
axes[1].axhline(gat_test_mae, color='red', linestyle='--', label=f'GNN teacher ({gat_test_mae:.3f} eV)')
axes[1].set_xlabel("Epoch"); axes[1].set_ylabel("Val MAE (eV)")
axes[1].set_title("Student vs GNN teacher accuracy")
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('phase4a_distillation.png', dpi=150, bbox_inches='tight')
plt.show()

torch.save(student.state_dict(), 'student_mlp.pt')
print("\n✅ Phase 4A complete! Student model saved as student_mlp.pt")
```
